```python
def HS4_to_HS2(element):
    
    element_start = re.match(r'(\d\d)\d\d', element).groups()[0]
    element_start = int(element_start)
     
    if (element_start >= 1 and element_start <= 5):
        new_element = 'Animal & Animal Products'
        
    elif (element_start >= 6 and element_start <= 15):
        new_element = 'Vegetable Products'
        
    elif (element_start >= 16 and element_start <= 24):
        new_element = 'Foodstuffs'
        
    elif (element_start >= 25 and element_start <= 27):
        new_element = 'Mineral Products'
        
    elif (element_start >= 28 and element_start <= 38):
        new_element = 'Chemicals & Allied Industries'
        
    elif (element_start >= 39 and element_start <= 40):
        new_element = 'Plastics / Rubbers'
        
    elif (element_start >= 41 and element_start <= 43):
        new_element = 'Raw Hides, Skins, Leather, & Furs'
        
    elif (element_start >= 44 and element_start <= 49):
        new_element = 'Wood & Wood Products'
        
    elif (element_start >= 50 and element_start <= 63):
        new_element = 'Textiles'
        
    elif (element_start >= 64 and element_start <= 67):
        new_element = 'Footwear / Headgear'
        
    elif (element_start >= 68 and element_start <= 71):
        new_element = 'Stone / Glass'
        
    elif (element_start >= 72 and element_start <= 83):
        new_element = 'Metals'
        
    elif (element_start >= 84 and element_start <= 85):
        new_element = 'Machinery / Electrical'
        
    elif (element_start >= 86 and element_start <= 89):
        new_element = 'Transportation'
        
    elif (element_start >= 90 and element_start <= 97):
        new_element = 'Miscellaneous'
        
    else:
        new_element = ''
    
    return new_element
```

```python
columns_types = {'year': np.int64,
                 'origin': str, 'hs07': str,
                 'export_val': np.float64, 'import_val': np.float64,
                 'export_rca': np.float64, 'import_rca': np.float64}
```

```python
country_names_df = pd.read_csv('country_names.tsv', sep = '\t')
products_names_df = pd.read_csv('products_hs_07.tsv', sep = '\t')
raw_data_df = pd.read_csv('year_origin_hs07_4.tsv', sep = '\t', dtype = columns_types)
```

```python
processed_data_df = raw_data_df.copy()
processed_data_df.drop(['import_val', 'export_rca', 'import_rca'], axis = 1, inplace = True)
processed_data_df.hs07 = processed_data_df.hs07.apply(HS4_to_HS2)
```
